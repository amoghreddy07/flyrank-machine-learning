```python
# My Lane

**Selected Lane:** Refresh / Content Opportunity Scoring

I selected the **Refresh / Content Opportunity Scoring** lane because it focuses on a practical business decision: identifying which webpages should be reviewed first for content updates based on observable search and engagement signals.

The goal is not simply to build a machine learning model, but to support content teams in prioritizing their work more effectively. A data-driven ranking can help reduce manual effort while making refresh decisions more consistent and explainable.

I chose this lane because it combines machine learning with a real business problem, making it both technically interesting and practically useful.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
## Decision
The decision is determining which webpages should be prioritized for content refresh instead of reviewing every page manually.

## Action
The content or SEO team can use the ranked list to update high-priority pages first, improving the efficiency of their workflow.

## Cost of a wrong recommendation
If the model incorrectly recommends a page for refresh, time and resources may be wasted on a page that doesn't need immediate attention. If it misses an important declining page, that page may continue losing search visibility and traffic before it is reviewed.

## Why ML helps
Many content signals such as impressions, CTR, content age, word count, and search performance interact together. Machine learning can combine these signals to produce a more consistent prioritization than simple manual rules.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))
print("Average CTR:", round(df["ctr"].mean(), 3))
print("Median Word Count:", int(df["word_count"].median()))

declining_pct = (df["trend_direction"] == "down").mean() * 100
print(f"Declining Pages: {declining_pct:.1f}%")

corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Search Volume vs Impressions Correlation: {corr:.3f}")
```
