*Classification, clustering, ranking, or scoring — which one, and why?*

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Target / Proxy

**Target (Prediction):**

The model will predict a **refresh priority score** for each webpage.

**Proxy Target:**

Since there is no direct "refresh priority" column, the model can use the existing decline label (or declining trend) as a proxy for identifying pages that are likely to benefit from a content refresh.

**Why a Proxy?**

The real business goal is to know which pages should be refreshed first. Because that decision is not directly stored in the dataset, we use the observed trend (such as whether a page is declining) as a proxy for refresh priority.
"""))
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
from IPython.display import Markdown, display

display(Markdown("""
### Success Metric

**Primary Success Metric: Precision@50**

Precision@50 measures how many of the top 50 pages recommended by the model actually require a content refresh.

**Why Precision@50?**

Content teams usually have limited time and resources. Instead of reviewing every page, they focus on the highest-priority pages first. A high Precision@50 means the model is successfully placing the most valuable refresh opportunities at the top of the ranking, helping teams spend their effort where it has the greatest impact.
"""))
```
